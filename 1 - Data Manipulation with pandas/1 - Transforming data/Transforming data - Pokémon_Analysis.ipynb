#**Subsetting rows**
Create a logical condition to filter the dataset.

````
pokemon[pokemon["HP"] > 200]
````

````
pokemon[pokemon['Name'] == 'Pikachu']
````

````
pokemon[(pokemon['HP'] > 100) & (pokemon['Attack'] > 100)].sort_values(['HP', 'Attack'], ascending = [False,False]).head()
# This code runs perfectly, but we can do better:
````

````
hp_over_100 = pokemon['HP'] > 100
attack_over_100 = pokemon['Attack'] > 100

(pokemon[hp_over_100 & attack_over_100]
 .sort_values(['HP', 'Attack'], ascending = [False, False])
 .head())
````

#**isin!!!!!**

````
color_list = ['Brown', 'Yellow']
pokemon[pokemon['Color'].isin(color_list)].head()
````

#**Adding a new column**

````
pokemon['Height_cm'] = pokemon['Height_m'] * 100
pokemon.head()
````

````
pokemon['BMI'] = pokemon['Weight_kg'] / pokemon['Height_m'] ** 2

pokemon.sort_values('BMI', ascending = True).head()
````

#**Multiple manipulations**

````
bmi_lt_100 = pokemon[pokemon['BMI'] < 100]
bmi_lt_100_height = bmi_lt_100.sort_values('Height_cm', ascending = False)
bmi_lt_100_height[['Name', 'Height_cm', 'BMI']]
````
